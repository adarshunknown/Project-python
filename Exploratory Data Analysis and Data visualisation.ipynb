Creating Date,Month and Year columns 

```python
days=list()#dividing the date into columns of days ,month and years
months=list()# the Date is divided into 3 parts, by making 3 empty lists , splitting them and appending them to the empty lists 
years=list()
for x in a["Date"]:
    x_split=x.split()
    days.append(x_split[1])
    months.append(x_split[0])
    years.append(x_split[2])
```

```python
a["Days"]=days
a["Months"]=months
a["Years"]=years
```

```python
a
```

```python
a1=a.isnull()
a1
```

```python
a["Battery Electric Vehicles (BEVs)"] = pd.to_numeric(a["Battery Electric Vehicles (BEVs)"].str.replace(",", "").str.replace(".",""), errors='coerce')
a["Plug-In Hybrid Electric Vehicles (PHEVs)"] = pd.to_numeric(a["Plug-In Hybrid Electric Vehicles (PHEVs)"].str.replace(",", "").str.replace(".",""), errors='coerce')
a["Electric Vehicle (EV) Total"] = pd.to_numeric(a["Electric Vehicle (EV) Total"].str.replace(",", "").str.replace(".",""), errors='coerce')
a["Non-Electric Vehicle Total"] = pd.to_numeric(a["Non-Electric Vehicle Total"].str.replace(",", "").str.replace(".",""), errors='coerce')
a["Total Vehicles"] = pd.to_numeric(a["Total Vehicles"].str.replace(",", "").str.replace(".",""), errors='coerce')
```

```python
a
```

```python
#The agg function in Pandas is a powerful tool for aggregating data within a DataFrame. It allows  to compute multiple statistics for one or more columns simultaneously.
```

```python
a1_county=a.groupby("County").agg({"Electric Vehicle (EV) Total":"sum","Total Vehicles":"sum"})
a1_county
```

```python
a1_county["Electric Vehicle Ratio"]=a1_county["Electric Vehicle (EV) Total"]/a1_county["Total Vehicles"]
```

```python
a1_county
a1_county["County"]=a1_county.index
a1_county
```

```python
a1_county=a1_county.sort_values(by="Electric_Vehicle_Ratio",ascending=False)
a1_county
```

```python
a2_county=a1_county.rename(columns={"Total Vehicles":"Total_Vehicles","Electric Vehicle Ratio":"Electric_Vehicle_Ratio"},inplace=True)
```
