creatimg a movie_matrix, to store user ratings such that
there are hundred users 
each user can review as many movies as they want
the review should be in between 0 and 10
the movies which are not reviewed are given a value of -1

```python
import random
movie_matrix=[]
for user in range(100):
    movies_rated_by_me=np.full(1000,-1)
    num_of_movies_rated=random.randint(0,999)
    movies_that_i_will_rate=random.sample(range(0,1000),num_of_movies_rated)
    for index in movies_that_i_will_rate:
        movies_rated_by_me[index]=random.randint(0,10)
    movie_matrix.append(movies_rated_by_me)
```

adding the reviwes of 10 movie experts and 50 more movies////

creating a matrix of 10 movie experts

```python
expert_matrix=[]
for user in range(10):
    movies_rated_by_me=np.full(1000,-1)
    num_of_movies_rated=random.randint(0,999)
    movies_that_i_will_rate=random.sample(range(0,1000),num_of_movies_rated)
    for index in movies_that_i_will_rate:
        movies_rated_by_me[index]=random.randint(0,10)
    expert_matrix.append(movies_rated_by_me)
expert_matrix=np.array(expert_matrix)
print(expert_matrix)
print(expert_matrix.size)
print(expert_matrix.shape)
print(type(expert_matrix))
print(len(expert_matrix))
```

```python
movie_matrix=np.vstack([movie_matrix,expert_matrix])
```

```python
movie_matrix.shape
```

adding 50 more movies 

```python
new_movie_matrix=[]
for user in range(110):
    movie_rated_by_me=np.full(50,-1)
    num_of_movies_rated=random.randint(0,49)
    movies_that_i_will_rate=random.sample(range(0,50),num_of_movies_rated)
    for index in movies_that_i_will_rate:
        movie_rated_by_me[index]=random.randint(0,10)
    new_movie_matrix.append(movie_rated_by_me)
new_movie_matrix=np.array(new_movie_matrix)
print(new_movie_matrix)
print(new_movie_matrix.shape)
print(new_movie_matrix.size)
print(type(new_movie_matrix))
```

```python
new_movies_id=np.append(movie_id,np.arange(2301,2351))
new_movies_id.size
```

```python
movie_matrix=np.hstack([movie_matrix,new_movie_matrix])
print(movie_matrix.shape)
```

creating final movie rating with four columns

```python
final_movie_rating=[]
for i in range(1050):
    m=movie_matrix[:,i]
    m=m[m>=0]
    total_num_rating=m.size
    avg_rating=m.mean()
    standard_deviation=m.std()
    final_movie_rating.append([new_movies_id[i],avg_rating,total_num_rating,standard_deviation])
final_movie_rating=np.array(final_movie_rating)
```

```python
final_movie_rating
```

convert the final movie ratings to have range from 0 to 10,such that the minimum ratings converts to 0 and maximum converts to 10,others values in between

```python
x=final_movie_rating[:,1]
old_range=x.max()-x.min()
new_range=(10-0)
final_movie_rating[:,1]=((x-x.min())*(new_range/old_range)+0)
```

```python
final_movie_rating
```

display the movies rating wise,highest to lowest

```python
sorted_films= final_movie_rating[final_movie_rating[:,1].argsort()[::-1]]
```
